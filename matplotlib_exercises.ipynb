### 1.
Use matplotlib to plot the following equation:
   y = x^2 - x + 2


```python
x_point = range(-50,50)
y = [x*x - x + 2 for x in x_point]
```

```python
plt.plot(x_point,y)

plt.title('$ y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.annotate('The Origin', xy = (0,0), xytext = (25,100), arrowprops = dict(facecolor = 'gray'))

# save the figure
plt.savefig('figure1.png')
```
